##### 1. 히스토그램: 데이터프레임의 열 데이터 분포와 빈도를 살펴보는 그래프

```python
axes1.hist(tips['total_bill'],bins=10) # bins: x축 간격 조정
axes1.set_title('Hisogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

fig
```

##### 2. 산점도 그래프: 변수 2개를 사용해서 만드는 그래프

- 이변량 그래프: 변수 두개를 사용하는 그래프 

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('Scatterplot of Totall Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

##### 3. 박스 그래프: 이산형 변수와 연속형 변수를 함께 사용하는 그래프

- 이산형 변수: Female, male과 같이 명확하게 구분되는 값

- 연속형 변수: Tip과 같이 명확하게 셀 수 없는 범위의 값

```python
boxplot=plt.figure()
axes1=boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'],
               tips[tips['sex']=='Male']['tip']],
              labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

### 다변량 그래프 그리기

####1. 산점도 그래프


```python
def recode_sex (sex):
  if sex=='Female':
    return 0
  else:
    return 1
```

```python
tips['sex_color']=tips['sex'].apply(recode_sex) #recode_sex가 반환한 값(0,1)을 df의 sex_colr에 추가
```

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    alpha=0.5
)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 4-3. seaborn 라이브러리 자유자재로 사용하기

### 1) 단변량 그래프 그리기

#### 히스토그램

```python
import seaborn as sns
tips=sns.load_dataset('tips')
```

```python
ax=plt.subplots()  #subplots()로 기본틀 생성
ax=sns.distplot(tips['total_bill'])   #distplot method에 total_bill 열 데이터 전달
ax.set_title('Total Bil Histogram with Density Plot')
```

```python
#distplot method 사용 시 히스토그램과 밀집도 그래프를 같이 그림
#밀집도 그래프 제외하고 싶다면 kde=False로 설정하면 됨

#밀집도 그래프란? 주어진 데이터를 정규화시켜 넓이가 1이 되도록 그린 그래프프
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], kde=False) 
ax.set_title('Total Bill Histogram') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Frequency')
```

```python
#밀집도 그래프만 나타내려면 hist 인자를 False로 지정하면 됨

ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], hist=False) 
ax.set_title('Total Bill Density') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Unit Probability')
```

```python
# distplot method에 rug 인자를 추가하여 양탄자 그래프 그리기
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
#count 그래프 그리기: tips df의 day열 데이터를 넣기
ax = plt.subplots() 
ax = sns.countplot(x = 'day',  data = tips) 
ax.set_title('Count of days') 
ax.set_xlabel('Day of the Week') 
ax.set_ylabel('Frequency') 
```

### 2) 이변량 그래프 그리기

## 4-5. seaborn 라이브러리로 그래프 스타일 설정하기

#### <1> 그래프에 스타일 적용하기

```python
fig, ax=plt.subplots()
ax=sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

#### <2> whitegrid: 그래프 배경에 가로줄 만들기

```python
sns.set_style('whitegrid')
fig, ax=plt.subplots()
ax=sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

#### <3> for문 이용하여 모든 스타일을 하나씩 적용

```python
fig=plt.figure()
seaborn_styles=['darkgrid','whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position=idx+1
  with sns.axes_style(style):
    ax=fig.add_subplot(2,3,plot_position)
    violin=sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
    violin.set_title(style)


fig.tight_layout()
```
